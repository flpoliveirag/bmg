```python
%sql

CREATE OR REPLACE TEMPORARY VIEW
  agg_beneficiarios
AS
  SELECT
    NM_RAZAO_SOCIAL,
    NM_MUNICIPIO,
    DE_FAIXA_ETARIA_REAJ,
    DE_SEGMENTACAO_PLANO,
    DE_ABRG_GEOGRAFICA_PLANO,
    SUM(QT_BENEFICIARIO_ATIVO) AS QT_BENEFICIARIO_ATIVO
  FROM
    refined.beneficiarios
  GROUP BY
    NM_RAZAO_SOCIAL,
    NM_MUNICIPIO,
    DE_FAIXA_ETARIA_REAJ,
    DE_SEGMENTACAO_PLANO,
    DE_ABRG_GEOGRAFICA_PLANO;
```

```python
%skip
%sql

CACHE TABLE agg_beneficiarios;
```

```python
%sql

SELECT count(*) AS N FROM agg_beneficiarios;
```


#### Q1. Quais são as **5 operadoras** com **maior número de beneficiários ativos**?

```python
%sql

SELECT
  NM_RAZAO_SOCIAL,
  sum(QT_BENEFICIARIO_ATIVO) AS QT_BENEFICIARIO_ATIVO
FROM
  agg_beneficiarios
GROUP BY
  NM_RAZAO_SOCIAL
ORDER BY
  QT_BENEFICIARIO_ATIVO DESC
LIMIT
  5;
```


#### Q2. Qual é a **faixa etária com mais beneficiários** e **quantos** são?

```python
%sql

SELECT
  DE_FAIXA_ETARIA_REAJ,
  sum(QT_BENEFICIARIO_ATIVO) AS QT_BENEFICIARIO_ATIVO
FROM
  agg_beneficiarios
GROUP BY
  DE_FAIXA_ETARIA_REAJ
ORDER BY
  QT_BENEFICIARIO_ATIVO DESC
LIMIT
  1;
```


#### Q3. Liste, de forma **decrescente** a **quantidade de beneficiários por município**.

```python
%sql

SELECT
  NM_MUNICIPIO,
  sum(QT_BENEFICIARIO_ATIVO) AS QT_BENEFICIARIO_ATIVO
FROM
  agg_beneficiarios
GROUP BY
  NM_MUNICIPIO
ORDER BY
  QT_BENEFICIARIO_ATIVO DESC;
```
