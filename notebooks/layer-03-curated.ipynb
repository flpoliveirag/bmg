- Finalmente, a camada curated foi criada usando como partições os campos relacionados à divisão de arquivos feita pela própria ANS.

<br>

```python
%sql

CREATE OR REPLACE TABLE
  curated.beneficiarios
USING
  DELTA
PARTITIONED BY
  (ID_CMPT_MOVEL, SG_UF)
FROM (
  SELECT
    * EXCEPT (NR_CNPJ, INGESTION_TS),
    current_timestamp() AS INGESTION_TS
  FROM
    refined.beneficiarios
);
```
