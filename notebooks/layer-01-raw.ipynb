```python
%sql

CREATE OR REPLACE TABLE
  raw.beneficiarios
AS
  SELECT
    * EXCEPT(_rescued_data),
    current_timestamp() AS INGESTION_TS
  FROM
    read_files(
      '/Volumes/workspace/default/ans/extracted/',
      format => 'csv',
      header => true,
      sep => ';',
      inferSchema => false
    );
```
