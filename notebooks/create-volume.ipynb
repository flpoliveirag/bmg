```python
%sh

url="https://dadosabertos.ans.gov.br/FTP/PDA/informacoes_consolidadas_de_beneficiarios-024/202508/pda-024-icb-TO-2025_08.zip"
filename=$(basename "$url")

vol="/Volumes/workspace/default/ans"
snk_zip="$vol/compressed/$filename"
snk_csv="$vol/extracted"

curl "$url" --output "$snk_zip"

unzip "$snk_zip" -d "$snk_csv" 
```
